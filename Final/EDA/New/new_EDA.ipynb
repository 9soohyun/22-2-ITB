```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

df1 = pd.read_csv('서울특별시 공공자전거 신규가입자 정보(일별)_2018년.csv', encoding='utf-8')
df2 = pd.read_csv('서울특별시 공공자전거 신규가입자 정보(일별)_18.07.20-21.12.31.csv', encoding='utf-8')
```

```python
df1
```

```python
df2
```

```python
df1['가입일자'] = pd.to_datetime(df1['가입일자'])
df1['가입일자'] = df1['가입일자'].dt.strftime('%Y-%m-%d')
df1 = df1.loc[:,['가입일자', '가입 수']]
df1
```

```python
df2['가입일자'] = '20' + df2['가입일자'].astype(str)
df2 = df2.loc[:,['가입일자', '가입 수']]
df2
```

```python
df = pd.concat([df1,df2])
df
```

```python
df = df.groupby(['가입일자'], as_index=False).sum()
df
```

```python
df.to_csv('./new.csv')
```
