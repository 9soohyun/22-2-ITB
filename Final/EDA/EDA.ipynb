```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
air = pd.read_csv('air.csv', encoding='utf-8')
air = air.drop(['Unnamed: 0'], axis=1)
air.rename(columns = {'측정일시': '일시'}, inplace = True)
air['일시'] = air['일시'].astype(str)
air['일시'] = pd.to_datetime(air['일시'])
air['일시'] = air['일시'].dt.strftime('%Y-%m-%d')
air
```

```python
new = pd.read_csv('new.csv', encoding='utf-8')
new = new.drop(['Unnamed: 0'], axis=1)
new.rename(columns = {'가입일자': '일시'}, inplace = True)
new['일시'] = new['일시'].astype(str)
new['일시'] = pd.to_datetime(new['일시'])
new['일시'] = new['일시'].dt.strftime('%Y-%m-%d')
new
```

```python
sub = pd.read_csv('subway.csv', encoding='utf-8')
sub = sub.drop(['Unnamed: 0'], axis=1)
sub.rename(columns = {'날짜': '일시'}, inplace = True)
sub['일시'] = sub['일시'].astype(str)
sub['일시'] = pd.to_datetime(sub['일시'])
sub['일시'] = sub['일시'].dt.strftime('%Y-%m-%d')
sub
```

```python
wea = pd.read_csv('weather_EDA.csv', encoding='utf-8')
wea = wea.drop(['Unnamed: 0'], axis=1)
wea.columns = ['일시', '강수량', '겨울기온', '풍속', '겨울체감', '여름기온', '습도', '여름체감']
wea.rename(columns = {'일자': '일시'}, inplace = True)
wea['일시'] = '20' + wea['일시'].astype(str)
wea
```

```python
sc = pd.read_csv('sc.csv', encoding='utf-8')
sc = sc.loc[:,['자치구 기준일', '기타 추가']]
sc.columns = ['일시', '확진자']
sc['일시'] = pd.to_datetime(sc['일시'])
sc['일시'] = sc['일시'].dt.strftime('%Y-%m-%d')
sc = sc[340:]
sc = sc.sort_values(by=["일시"], ascending=[True]) 
sc
```

```python
pd.set_option('display.max_seq_items', None)
value = sc['일시'].values
print(value)
```

```python
sc = sc[~sc.duplicated()]
sc
```

```python
df = pd.merge(air,sub, how='outer',on='일시')
df = pd.merge(df,wea, how='outer',on='일시')
df = pd.merge(df,new, how='outer',on='일시')
df = pd.merge(df,sc, how='outer',on='일시')
df
```

```python
df = df.fillna(0)
```

```python
df.to_csv('./data.csv')
```
