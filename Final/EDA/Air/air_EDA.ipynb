```python
import os
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
path = './/'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.csv')]
file_list_py
```

```python
df = pd.DataFrame()

for i in file_list_py:
    data = pd.read_csv(i, encoding='euc-kr')
    data.columns = ['측정일시', '권역명', '측정소명', '이산화질소', '오존',
                    '일산화가스','아황산가스', '미세먼지', '초미세먼지'
                   ]
    df = pd.concat([df,data])
```

```python
df = df.loc[:,['측정일시','측정소명', '오존', '미세먼지']]
df
```

```python
list = ['광진구']
df_k = df[df['측정소명'].isin(list)]
df_k = df_k.drop(['측정소명'], axis=1)
df_k = df_k.reset_index(drop=True)
df_k.columns = ['측정일시','광진구오존', '광진구미세먼지']
df_k
```

```python
list = ['동대문구']
df_d = df[df['측정소명'].isin(list)]
df_d = df_d.drop(['측정소명'], axis=1)
df_d = df_d.reset_index(drop=True)
df_d.columns = ['측정일시','동대문구오존', '동대문구미세먼지']
df_d
```

```python
list = ['성동구']
df_s = df[df['측정소명'].isin(list)]
df_s = df_s.drop(['측정소명'], axis=1)
df_s = df_s.reset_index(drop=True)
df_s.columns = ['측정일시','성동구오존', '성동구미세먼지']
df_s
```

```python
list = ['중랑구']
df_j = df[df['측정소명'].isin(list)]
df_j = df_j.drop(['측정소명'], axis=1)
df_j = df_j.reset_index(drop=True)
df_j.columns = ['측정일시','중랑구오존', '중랑구미세먼지']
df_j
```

```python
df = pd.merge(df_k, df_d, how='outer',on='측정일시')
df = pd.merge(df, df_s, how='outer',on='측정일시')
df = pd.merge(df, df_j, how='outer',on='측정일시')
df
```

```python
df.to_csv('./air.csv')
```
