```python
import os
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
path = './/'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.csv')]
```

```python
file_list_py
```

```python
df = pd.DataFrame()

for i in file_list_py:
    data = pd.read_csv(i, encoding='euc-kr')
    data.columns = ['날짜', '노선', '역명', '승차', '하차', '등록일']
    df = pd.concat([df,data])

df = df.drop(['등록일', '노선'], axis=1)
```

```python
list = ['강변(동서울터미널)', '건대입구', '광나루(장신대)', '구의(광진구청)', '군자(능동)',
       '아차산(어린이대공원후문)', '중곡', '어린이대공원(세종대)', '뚝섬유원지']
df_k = df[df['역명'].isin(list)]
df_k[['승차']] = df_k[['승차']].apply(pd.to_numeric)
df_k[['하차']] = df_k[['하차']].apply(pd.to_numeric)
df_k['광진구유동인구'] = (df_k['승차'] + df_k['하차'])
df_k = df_k.drop(['승차', '하차', '역명'], axis=1)
df_k = df_k.groupby(['날짜'], as_index=False).sum()
df_k
```

```python
list = ['신설동', '제기동', '청량리(서울시립대입구)', '용두(동대문구청)', '장한평']
df_d = df[df['역명'].isin(list)]
df_d[['승차']] = df_d[['승차']].apply(pd.to_numeric)
df_d[['하차']] = df_d[['하차']].apply(pd.to_numeric)
df_d['동대문구유동인구'] = (df_d['승차'] + df_d['하차'])
df_d = df_d.drop(['승차', '하차', '역명'], axis=1)
df_d = df_d.groupby(['날짜'], as_index=False).sum()
df_d
```

```python
list = ['금호', '답십리', '상왕십리', '왕십리(성동구청)', '한양대', '뚝섬', '성수', '용답', '신답',
       '옥수', '신금호', '행당', '마장']
df_s = df[df['역명'].isin(list)]
df_s[['승차']] = df_s[['승차']].apply(pd.to_numeric)
df_s[['하차']] = df_s[['하차']].apply(pd.to_numeric)
df_s['성동구유동인구'] = (df_s['승차'] + df_s['하차'])
df_s = df_s.drop(['승차', '하차', '역명'], axis=1)
df_s = df_s.groupby(['날짜'], as_index=False).sum()
df_s
```

```python
list = ['봉화산(서울의료원)', '신내', '먹골', '면목', '사가정', '용마산(용마폭포공원)']
df_j = df[df['역명'].isin(list)]
df_j[['승차']] = df_j[['승차']].apply(pd.to_numeric)
df_j[['하차']] = df_j[['하차']].apply(pd.to_numeric)
df_j['중랑구유동인구'] = (df_j['승차'] + df_j['하차'])
df_j = df_j.drop(['승차', '하차', '역명'], axis=1)
df_j = df_j.groupby(['날짜'], as_index=False).sum()
df_j
```

```python
df = pd.merge(df_k, df_d, how='outer',on='날짜')
df = pd.merge(df, df_s, how='outer',on='날짜')
df = pd.merge(df, df_j, how='outer',on='날짜')
df
```

```python
df.to_csv('./subway2018.csv')
```
