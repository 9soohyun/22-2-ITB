```python
import os
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
path = './/'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('subway.csv')]
```

```python
file_list_py
```

```python
df = pd.DataFrame()

for i in file_list_py:
    data = pd.read_csv(i, encoding='utf-8')
    df = pd.concat([df,data])
```

```python
df
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
df
```

```python
df.to_csv('./subway.csv')
```
